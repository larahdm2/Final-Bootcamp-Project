##### Request

> This web cannot be read with Beautifulsoup. It could be read with Selenium, but for time management reasons, I have copied the table in a text file by selecting 'Edit as HTML'.

```python
file = open('fbp_html.txt')
```

```python
response = file.read()
```

```python
response
```

##### Soup

```python
soup = BeautifulSoup(response, "html.parser")
```

```python
soup
```

##### Select

```python
soup.select('td')
```

```python
soup.select('td')[1103].get_text()
```

##### List

```python
population = []

for i in range(0, 1104, 6):
    population.append(soup.select('td')[i].get_text())
```

```python
number = []

for i in range(1, 1104, 6):
    number.append(soup.select('td')[i].get_text())
```

##### Dataframe

```python
data = pd.DataFrame({"population":population, "number":number})
```

```python
data
```

##### Save

```python
data.to_csv('fbp_data1.csv', index=False)
```
